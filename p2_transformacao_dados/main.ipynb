## Projeto - Transformação de dados

Importação das bibliotecas necessárias

```python
import pandas as pd
import tabula
import zipfile
import os
```

### Passo a passo

### 1. Extrair os dados da tabela Rol de Procedimentos e Eventos em Saúde do PDF do Anexo I do projeto 1 (todas as páginas)

```python
# Caminho para o PDF "Anexo I"
anexo_1 = "C:/Users/jamil/OneDrive/Área de Trabalho/IntuitiveCare/p1_web_scraping/downloads/Anexo_I_Rol_2021RN_465.2021_RN627L.2024.pdf"

# Extração das tabelas do PDF - todas as páginas
tabelas = tabula.read_pdf(anexo_1, pages="all", multiple_tables=True)

# Verificando quantas tabelas foram extraídas
print(f"Total de tabelas extraídas: {len(tabelas)}")
```

```python
# Exibir as primeiras linhas das 3 primeiras tabelas extraídas
for i in range(min(3, len(tabelas))):  # Garante que não dá erro se tiver menos de 3 tabelas
    print(f"\n=== Tabela {i+1} ===")
    display(tabelas[i].head())  # Exibe as primeiras linhas da tabela
```

```python
# Remover a primeira tabela da lista devido as informações serem irrelevantes
tabelas = tabelas[1:]

# Verificando se a tabela foi removida
print(f"Total de tabelas restantes: {len(tabelas)}")
```

Concatenando todas as tabelas em um único DataFrame

```python
# Concatenar todas as tabelas extraídas em um único DataFrame
df_total = pd.concat(tabelas)

# Visualizar as primeiras linhas do DataFrame final
display(df_total.head())
```

### 2. Substitua as abreviações das colunas OD (Odontológica) e AMB (Ambulatorial) pelas descrições completas.

```python
# Substituindo as siglas OD e AMB nas colunas da tabela por seus significados completos
df_total.replace({"OD": "Odontológica", "AMB": "Ambulatorial"}, inplace=True)

# Exibição das primeiras linhas para conferir a substituição se esta correta.
display(df_total.head())
```

### 3. Salve os dados em uma tabela estruturada, em formato csv.

```python
# Caminho para salvar o CSV
csv_path = "Teste_Jamil.csv"

# Salvando o CSV de forma organizada
df_total.to_csv(csv_path, index=False, encoding="utf-8", sep=";")



print(f"Arquivo CSV salvo em: {csv_path}")
```

### 4. Compacte o csv em um arquivo denominado "Teste_{seu_nome}.zip".

```python
# Nome do arquivo ZIP
zip_path = "Teste_Jamil.zip"

# Criando o ZIP e adicionando o arquivo CSV dentro dele
with zipfile.ZipFile(zip_path, "w", zipfile.ZIP_DEFLATED) as zipf:
    zipf.write(csv_path)

print(f"Arquivo ZIP criado: {zip_path}")
```
